```python
import requests as rq
import json
import pandas as pd
```

```python
FANTASYID = 598
```

```python
fantasy = rq.get(f"https://api.csapi.de/fantasy/{FANTASYID}").text

parsed_fantasy = json.loads(fantasy)


parsed_fantasy
```

```python
players = []


for team in parsed_fantasy['teams']:
    for player in team['players']:
        player_stats = rq.get(f"https://api.csapi.de/players/{player['id']}/stats/events?start_date=2026-02-20&mapid=0").text
        parsed_stats = json.loads(player_stats)[0]
        players.append({'player': player['name'], 'rating': parsed_stats['rating']})
        

```

```python
players
```

```python
df = pd.DataFrame.from_dict(players)
```

```python
df.to_csv('out.csv',index=False, header=False)

display(df)
```
